```python
# Ex 15.17 Binary Classification with the Breast Cancer Dataset on page 661.
# Sections: loading, sizes, split, training, score, confusion matrix, reports,
#           visualization, k-fold CV, and model comparison.

import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import (
    confusion_matrix,
    classification_report,
    ConfusionMatrixDisplay,
)

def main():
    # -----------------------------
    # Loading the Dataset
    # -----------------------------
    iris = load_iris()
    X, y = iris.data, iris.target
    target_names = iris.target_names
    feature_names = iris.feature_names

    print("=== Loading the Dataset ===")
    print(f"Feature names: {feature_names}")
    print(f"Target names: {list(target_names)}")

    # -----------------------------
    # Checking the Sample and Target Sizes
    # -----------------------------
    print("\n=== Checking the Sample and Target Sizes ===")
    print(f"Samples (X) shape: {X.shape}")       # (150, 4)
    print(f"Targets (y) shape: {y.shape}")       # (150,)

    # -----------------------------
    # Splitting the Data for Training and Testing
    # -----------------------------
    TEST_SIZE = 0.25
    RANDOM_STATE = 11

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y
    )

    # -----------------------------
    # Training and Testing Set Sizes
    # -----------------------------
    print("\n=== Training and Testing Set Sizes ===")
    print(f"X_train: {X_train.shape}, y_train: {y_train.shape}")
    print(f"X_test:  {X_test.shape}, y_test:  {y_test.shape}")

    # -----------------------------
    # Creating the Model (GaussianNB)
    # -----------------------------
    print("\n=== Creating the Model: GaussianNB ===")
    nb = GaussianNB()

    # -----------------------------
    # Training the Model (GaussianNB)
    # -----------------------------
    nb.fit(X_train, y_train)

    # -----------------------------
    # Predicting (GaussianNB)
    # -----------------------------
    nb_pred = nb.predict(X_test)

    # -----------------------------
    # Estimator Method score (GaussianNB)
    # -----------------------------
    nb_score = nb.score(X_test, y_test)
    print(f"GaussianNB test accuracy (score): {nb_score:.4f}")

    # -----------------------------
    # Confusion Matrix (GaussianNB)
    # -----------------------------
    nb_cm = confusion_matrix(y_test, nb_pred)
    print("\nGaussianNB Confusion Matrix:\n", nb_cm)

    # -----------------------------
    # Classification Report (GaussianNB)
    # -----------------------------
    print("\nGaussianNB Classification Report:")
    print(classification_report(y_test, nb_pred, target_names=target_names))

    # -----------------------------
    # Creating the Model (KNN default k=5)
    # -----------------------------
    print("\n=== Creating the Model: KNeighborsClassifier (default k=5) ===")
    knn = KNeighborsClassifier()  # default n_neighbors=5

    # -----------------------------
    # Training the Model (KNN)
    # -----------------------------
    knn.fit(X_train, y_train)

    # -----------------------------
    # Predicting (KNN)
    # -----------------------------
    knn_pred = knn.predict(X_test)

    # -----------------------------
    # Estimator Method score (KNN)
    # -----------------------------
    knn_score = knn.score(X_test, y_test)
    print(f"KNN (k=5) test accuracy (score): {knn_score:.4f}")

    # -----------------------------
    # Confusion Matrix (KNN)
    # -----------------------------
    knn_cm = confusion_matrix(y_test, knn_pred)
    print("\nKNN Confusion Matrix:\n", knn_cm)

    # -----------------------------
    # Classification Report (KNN)
    # -----------------------------
    print("\nKNN Classification Report:")
    print(classification_report(y_test, knn_pred, target_names=target_names))

    # -----------------------------
    # Visualizing the Confusion Matrix (KNN)
    # -----------------------------
    print("\n=== Visualizing the Confusion Matrix (KNN) ===")
    disp = ConfusionMatrixDisplay(confusion_matrix=knn_cm, display_labels=target_names)
    disp.plot(cmap="Blues")
    plt.title("Iris - KNN Confusion Matrix")
    plt.tight_layout()
    plt.show()

    # -----------------------------
    # K-Fold Cross-Validation
    # -----------------------------
    print("\n=== K-Fold Cross-Validation (n_splits=10, shuffle=True, random_state=11) ===")
    kfold = KFold(n_splits=10, random_state=RANDOM_STATE, shuffle=True)

    nb_cv_scores = cross_val_score(estimator=nb, X=X, y=y, cv=kfold)
    knn_cv_scores = cross_val_score(estimator=knn, X=X, y=y, cv=kfold)

    print(f"GaussianNB CV: mean={nb_cv_scores.mean():.4f}, std={nb_cv_scores.std():.4f}")
    print(f"KNN (k=5) CV:   mean={knn_cv_scores.mean():.4f}, std={knn_cv_scores.std():.4f}")

    # -----------------------------
    # Running Multiple Models to Find the Best One
    # -----------------------------
    print("\n=== Running Multiple Models to Find the Best One ===")
    estimators = {
        "GaussianNB": GaussianNB(),
        "KNeighborsClassifier": KNeighborsClassifier(),
        # LogisticRegression fix: wrap in OneVsRestClassifier (avoids deprecated multi_class)
        "LogisticRegression (OvR)": OneVsRestClassifier(
            LogisticRegression(solver="lbfgs", max_iter=10000)
        ),
        "SVC": SVC(gamma="scale"),
    }

    summary = []
    for name, est in estimators.items():
        # Fit on train
        est.fit(X_train, y_train)
        # Test accuracy via .score()
        test_acc = est.score(X_test, y_test)
        # 10-fold CV on the full dataset
        cv_scores = cross_val_score(estimator=est, X=X, y=y, cv=kfold)
        cv_mean, cv_std = cv_scores.mean(), cv_scores.std()
        summary.append((name, test_acc, cv_mean, cv_std))
        print(
            f"{name:>25} | Test Acc: {test_acc:.4f} | CV Mean: {cv_mean:.4f} | CV Std: {cv_std:.4f}"
        )

    # Sort by test accuracy, then CV mean
    summary_sorted = sorted(summary, key=lambda t: (t[1], t[2]), reverse=True)

    print("\n=== Model Ranking (by Test Accuracy, then CV Mean) ===")
    for rank, (name, test_acc, cv_mean, cv_std) in enumerate(summary_sorted, start=1):
        print(
            f"{rank}. {name:>25} | Test Acc: {test_acc:.4f} | CV Mean: {cv_mean:.4f} | CV Std: {cv_std:.4f}"
        )

    # Final answer for the assignment prompt (KNN accuracy)
    print("\n=== Final Answer ===")
    print(f"KNeighborsClassifier (k=5) Test Accuracy: {knn_score:.4f}")

if __name__ == "__main__":
    main()
```
